```python
#This will convert 200 minutes to hours by diving by 60
def minutos_a_horas(minutos):
  horas = minutos / 60
  return horas

# Ejemplo de uso
minutos_totales = 200
horas_resultantes = minutos_a_horas(minutos_totales)
print(f"{minutos_totales} minutos son {horas_resultantes} horas.")
```
